### Calcular Rw- a partir de la curva de SP

La resistividad del agua (Rw) puede igualmente ser calcular a partir del registro de potencial espontáneo (SP). Dicha ecuación necesita algunos valores como la resisitividad del filtrado del lodo a temperatura de formación, la medición del registro SP y una constante K dependiente de la la temperatura de formación. La fórmula es la siguiente:

```python
import math
from math import log
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pozo = pd.read_csv(r'C:\Users\Usuario\Archivos_LAS\plot-data.csv')
```

Estos son los datos

```python
pozo
```

**Despliegue de cada curva en un solo carril**

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 10), sharey=True)

# 1er eje: Solo curva SP

ax[0].plot(pozo['SP'], pozo['DEPTH'], color='brown', label='SP')
ax[0].set_xlim(-160, 40)
ax[0].set_xlabel('SP (mV)', color='brown')
ax[0].set_xticks([-160, -100, -40, 0, 40])
ax[0].tick_params(axis='x', colors='brown')
ax[0].grid(True, which='both')

# 2do eje: IL y SN (resistividad)

ax[1].plot(pozo['IL'], pozo['DEPTH'], color='blue', label='IL')
#ax[1].set_xscale('log')
ax[1].set_xlim(0, 50)
ax[1].set_xlabel('IL (Ohm·m)', color='blue')
#ax[1].set_xticks([0.2, 1, 10, 50])
ax[1].get_xaxis().set_major_formatter(plt.ScalarFormatter())  # Mostrar valores log
ax[1].tick_params(axis='x', colors='blue')
ax[1].grid(True, which='both')

#ax2b = ax2.twiny()
ax[2].plot(pozo['SN'], pozo['DEPTH'], color='red', label='SN')
#ax[2].set_xscale('log')
ax[2].set_xlim(0, 50)
ax[2].set_xlabel('SN (Ohm·m)', color='red')
#ax[2].set_xticks([0.2, 1, 10, 50])
ax[2].get_xaxis().set_major_formatter(plt.ScalarFormatter())
ax[2].tick_params(axis='x', colors='red')
ax[2].grid(True, which='both')

ax[0].legend(loc='upper right', fontsize=12, framealpha=0.8)

plt.show()
```
