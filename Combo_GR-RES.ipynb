```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import lasio
```

### Leer archivo LAS

###  Tabular los datos del archivo LAS con pandas
Esto nos facilitará el manejo de los datos al tabularlos además *.reset_index()* asegura que 'DEPT' esté como columna

```python
well = las.df().reset_index() 
depth = well.index
```

### Graficar las curvas
Este bloque de código lo podemos subdividir en tres pasos

 * Configuración geométrica, vamos a representar una curva en cada track o carril es necesario definir subgráficos,            entonces la figura contendrá dos ejes principales ax1 y ax2 con tamaño indicado. 
 * Una configuración de los ejes, por ser simplemente una curva entonces colspan =1 entonces la matriz es (1,2) para cada      eje
 * La tercera parte corresponde a todos los parámetros que definen la visualización de la curva

```python
#################
top_depth = 3500
bot_depth = 4150
##################

fig, ax = plt.subplots(figsize=(7, 8))
fig.suptitle(f"Well:{'Pozo tipo'}\n( Interval: {top_depth} - {bot_depth} )", 
             fontsize=12, x=0.65, y=1.00, horizontalalignment='right', verticalalignment='top')

################################ % ###############################

# Configuración de los ejes base
ax1 = plt.subplot2grid((1, 2), (0, 0), colspan=1)
ax2 = plt.subplot2grid((1, 2), (0, 1), colspan=1)

# ✅ Ocultar ticks y etiquetas del eje x inferior
for ax_base in [ax1, ax2]:
    ax_base.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

# ✅ Activar grid horizontal (profundidades) sobre ejes base
for ax_base in [ax1, ax2]:
    ax_base.yaxis.grid(True, which='major', linestyle='--', linewidth=0.5, color='gray')
    
################################ % ##################################

# Eje superior para GR (twiny)
gr = ax1.twiny()
gr.plot(well['GR'], well.index, color='green', linewidth=0.5)
gr.xaxis.set_minor_locator(AutoMinorLocator())
gr.tick_params(which='both', width=2)
gr.tick_params(which='major', length=7)
gr.tick_params(which='minor', length=4, color='r')
gr.set_xlabel('GR', fontsize=10)
gr.set_xlim(0, 150)
gr.set_ylim(bot_depth, top_depth)
gr.xaxis.label.set_color('green')
gr.xaxis.grid(which='minor', linestyle=':', linewidth='0.5', color='blue')
gr.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

# Eje superior para RES_DEP (twiny)
res = ax2.twiny()
res.plot(well["ILD"], well.index, color='red', linewidth=0.5)
res.xaxis.set_minor_locator(AutoMinorLocator())
res.tick_params(which='both', width=2)
res.tick_params(which='major', length=7)
res.tick_params(which='minor', length=4, color='r')
res.set_xlabel('RES_DEP', fontsize=10)
res.set_xlim(0.2, 2000)
res.set_ylim(bot_depth, top_depth)
res.semilogx()
res.xaxis.grid(which='minor', linestyle=':', linewidth='0.5', color='blue')   
res.xaxis.label.set_color('red')
res.tick_params(axis='x', colors='red')
res.grid(which='major', color='lightgrey', linestyle='-')
res.tick_params(axis='x', which='both', bottom=False, labelbottom=False)



plt.show()
```
